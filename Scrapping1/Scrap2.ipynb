```python
for i in range (4):
    url = f"https://wuzzuf.net/search/jobs/?a=hpb&q=ai&start={i}"
    website = requests.get(url).text
    mywebsite = BeautifulSoup(website , "html")
    jobs = mywebsite.find_all("div" , class_="css-1gatmva e1v1l3u10")
    for job in jobs :
        jobtitle = job.find("h2" , class_="css-m604qf").a.text
        jobdesc = job.find("div" , class_="css-d7j1kk").text
        print(jobtitle)
        print(jobdesc)
        print("\n")
    
    
```

```python
import csv
```

```python
csvfile = open("AI.csv", "w" , encoding="utf-8")
writer = csv.writer(csvfile)
```

```python
for i in range (4):
    url = f"https://wuzzuf.net/search/jobs/?a=hpb&q=ai&start={i}"
    website = requests.get(url).text
    mywebsite = BeautifulSoup(website , "html")
    jobs = mywebsite.find_all("div" , class_="css-1gatmva e1v1l3u10")
    for job in jobs :
        jobtitle = job.find("h2" , class_="css-m604qf").a.text
        jobdesc = job.find("div" , class_="css-d7j1kk").text
        writer.writerow([jobtitle , jobdesc])
       
```
